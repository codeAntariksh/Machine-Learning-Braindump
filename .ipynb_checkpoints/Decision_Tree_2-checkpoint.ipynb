```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

iris = datasets.load_iris()
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
X_train = iris.data
y_train = iris.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train,test_size=0.2,random_state=23)
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X_train,y_train)
y_test = clf.predict(X_test)
tree.plot_tree(clf)
```

```python
feature_names = [
    "sepal length (cm)",
    "sepal width (cm)",
    "petal length (cm)",
    "petal width (cm)"
]

from dtreeviz import model

viz = model(
    clf,
    X_train,
    y_train,
    feature_names=feature_names,
    class_names=["setosa", "versicolor", "virginica"]
)

viz.view()

```

```python
tree.plot_tree(
    clf,
    feature_names=feature_names,   # 4 real feature names
    class_names=iris.target_names, # 3 real classes
    filled=True,
    rounded=True
)

```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, tree
from sklearn.model_selection import train_test_split
from dtreeviz import model

# Load Iris dataset
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=23
)

# Train Decision Tree Classifier
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# ------------------------
# 1 Scatter plot of first two features
# ------------------------
plt.figure(figsize=(8, 6))
scatter = plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Iris Dataset: Sepal Length vs Sepal Width")
handles, _ = scatter.legend_elements()
plt.legend(handles=handles, labels=iris.target_names.tolist(), title="Classes")
plt.show()


# ------------------------
# 2 Sklearn tree plot inline
# ------------------------
plt.figure(figsize=(12, 8))
tree.plot_tree(
    clf,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.title("Decision Tree Classification Plot")
plt.show()

# ------------------------
# 3 dtreeviz interactive visualization inline
# ------------------------
# If using Jupyter Notebook, you can render inline
viz = model(
    clf,
    X_train,
    y_train,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
)

# Inline rendering for Jupyter
viz.view()
```
